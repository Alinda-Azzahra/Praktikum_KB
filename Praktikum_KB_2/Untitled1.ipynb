# MEMBUAT DICTIONARY DENGAN KEY SESUAI NAMA INFORMASINYA

```python
total_record = len(df)
nama_attribute = df.columns
jumlah_objek = len(df.select_dtypes(include=['object']).columns)
jumlah_angka = len(df.select_dtypes(include=['int64', 'float64']).columns)
jumlah_NULL = len(df.columns[df.isna().sum() > 0])

dictionary_ps = {
    "Total Record": total_record,
    "Nama Attribute": nama_attribute,
    "Jumlah Objek": jumlah_objek,
    "Jumlah Angka": jumlah_angka,
    "Jumlah NULL": jumlah_NULL,
}

pd.Series(dictionary_ps)
```
